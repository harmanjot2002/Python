```python
#iterating over elements
a=[[2,4,6,8,10],[3,6,9,12,15],[4,8,12,16,20]]
for i in a:
    for j in i:
        print(j,end=" ")
    print()
```

### sum of all numbers in 2-D list

```python
#sum by iterating over elements
a=[[1,2,3,4],[5,6],[7,8,9]]
s=0
for i in a:
    for j in i:
        s=s+j
print(s)
```

```python
#sum by iterating over indices
a=[[1,2,3,4],[5,6],[7,8,9]]
s=0
for i in range(len(a)):
    for j in range(len(a[i])):
        s=s+a[i][j]
print(s)
```

```python
fruits = [["Banana", "Apple", "Watermelon", "Apple"], ["Apple", "Banana", "Watermelon", "Lemon"]]
print(fruits)

findFruit = input("Fruit: ")
replaceFruit = input("Replace: ")

for i in range(len(fruits)):    
    for j in range(len(fruits[i])):    
        if fruits[i][j] == findFruit:
            fruits[i][j] = replaceFruit


print(fruits)
```
